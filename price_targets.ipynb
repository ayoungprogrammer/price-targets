```python
import pandas as pd
import numpy as np

df = pd.read_csv('targets.csv')

df
```

```python
df['target_perc'] = (df['target'] - df['aver_close_at_analysis'])/df['aver_close_at_analysis'] * 100
df['real_perc'] = (df['aver_close_at_12m'] - df['aver_close_at_analysis'])/df['aver_close_at_analysis'] * 100
df['error'] = (df['target_perc']-df['real_perc'])
df['abs_error'] = abs(df['error'])

df[['ticker', 'analyst', 'target', 'aver_close_at_12m', 'target_perc', 'real_perc', 'error']]
```

```python
def filter_tail(data, p1=10, p2=90):
    q1 = np.percentile(data, p1)
    q3 = np.percentile(data, p2)
    return data[(data > q1) & (data < q3)]

analysts = df.groupby(['analyst'], as_index=False)['error'].agg({
        'mean_abs_err': lambda xs:np.mean(np.abs(filter_tail(xs))),
        'count': 'count',
        '10p': lambda xs: np.percentile(xs, q=10),
        '90p': lambda xs: np.percentile(xs, q=90),
        'mean': lambda xs: np.mean(filter_tail(xs)),
        'std': lambda xs: np.std(filter_tail(xs)),
    })
analysts = analysts[analysts['count'] > 100].sort_values(by='mean')
analysts[['analyst','std', 'mean', 'mean_abs_err', '10p', '90p', 'count']]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

ax = plt.axes()
ax.set_ylim(-100, 100)


xlabels = analysts['analyst']
x = [i for i in range(1, len(xlabels)+1)]
print(x)
y = analysts['mean_no_outliers']
e = analysts['std_no_outliers']
plt.errorbar(x, y, e, fmt='o')
plt.xlim(0, len(x) + 1)
plt.ylim(-50, 50)
plt.xticks(x, xlabels, rotation='vertical')
plt.axhline(y=0, xmin=0, xmax=10, ls='dashed', color='r')


# for name, analyst in analyst_error.items():
#     analyst['name'] = name

# analysts = [analyst for analyst in analyst_error.values() if len(analyst['diffs']) > 100]

# scotia = [analyst['mdiffs'] for analyst in analysts]
# plt.boxplot(scotia, labels=[analyst['name'] for analyst in analysts])
# plt.xticks(rotation='vertical')
plt.show()
```

```python
ratings = df[(df['rating'] != 'NaN') & (df['analyst'].isin(analysts['analyst']))]
ratings_agg = ratings.groupby(['analyst', 'rating'], as_index=False).agg({
        'error': {
            'mae': lambda xs: np.mean(np.abs(filter_tail(xs))),
        },
        'real_perc': {
            'mean': 'mean',
            'median': 'median',
            '10p': lambda xs: np.percentile(xs, 10),
            '90p': lambda xs: np.percentile(xs, 90),
            'count': 'count',
        },
        'target_perc': {
            'median': 'median',
            '10p': lambda xs: np.percentile(xs, 10),
            '90p': lambda xs: np.percentile(xs, 90),
        }
    })

ratings_agg.columns = list(map('_'.join, ratings_agg.columns.values))
ratings_agg[ratings_agg['real_perc_count'] > 10]
```

```python
top_ratings = ratings_agg[ratings_agg['real_perc_count'] > 20]
top_ratings = top_ratings.sort_values('real_perc_mean', ascending=False)
top_ratings['analyst_rating'] = top_ratings['analyst_'] + ' ' + top_ratings['rating_']
top_analyst_ratings = top_ratings.head(10)
top_analyst_ratings
```

```python
#ax = plt.axes()
#ax.set_ylim(-100, 100)
df['analyst_rating'] = df['analyst'] + ' '+ df['rating']
ratings_perf = df[df['analyst_rating']
                .isin(top_analyst_ratings['analyst_rating'])]
                .groupby('analyst_rating',as_index=False)['real_perc'].agg({
                    'mean': 'mean',
                    'std': 'std',
                }).sort_values(by='mean')
#ratings_perf
#ratings_perf.boxplot(ax=ax, by='analyst_rating', column='real_perc', return_type='axes', rot=90)
xlabels = ratings_perf['analyst_rating']
x = [i for i in range(1, len(xlabels)+1)]
print(x)
y = ratings_perf['mean']
e = ratings_perf['std']
plt.errorbar(x, y, e, fmt='o')
plt.xlim(0, len(x) + 1)
plt.ylim(-50, 150)
plt.xticks(x, xlabels, rotation='vertical')
```
